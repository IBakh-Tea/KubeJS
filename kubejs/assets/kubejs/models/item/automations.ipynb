### Bow and Crossbow (with Pistol and Musket) creation

```python
from shutil import copyfile

for type in ["", "long", "cross"]:
    for stave in custom_weapons["handle_materials"]:
        copyfile(f"template/{type}bow.json", f"weapons/{stave}_{type}bow.json")

for type in ["pistol", "musket", "arbalest"]:
    for stave in custom_weapons["handle_materials"]:
        for stock in custom_weapons["main_materials"]:
            if stock == "bone":
                continue
            copyfile(f"template/{type}.json", f"weapons/{stave}_{stock}_{type}.json")
```
